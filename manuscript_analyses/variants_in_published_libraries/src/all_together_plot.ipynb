Plot all the library analysis results together for the figure.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_context('paper')
sns.set_palette('colorblind')

%matplotlib inline
```

```python
bassik = pd.read_csv('../dat/bassik/1kg_wtc_bassik_results.tsv', sep='\t')
bassik['Library'] = 'Bassik'
brunello = pd.read_csv('../dat/brunello/1kg_wtc_brunello_results.tsv', sep='\t')
brunello['Library'] = 'Brunello'
sl = pd.read_csv('../dat/sabatini_lander/1kg_wtc_sabatini_lander_results.tsv', sep='\t')
sl['Library'] = 'Sabatini Lander'
```

```python
df_overall = pd.concat([bassik, brunello, sl])
```

```python
p = sns.catplot('n_vars', kind='count', row='Dataset', hue='Library', 
                data=df_overall, height=1.6, aspect=3.25, sharex=False,
               legend=False, sharey=True)
(p.set_axis_labels('Number of Variants', 'Number\n of sgRNAs\n (log scale)')
 .set_titles('{row_name}', verticalalignment='top'))
p.fig.get_axes()[0].set_yscale('log')
p.fig.suptitle('Variants in commonly used gRNA libraries')
plt.legend(bbox_to_anchor=(1.0, 1.0),loc=5)

p.savefig('../figs/grna_libraries_all_wtc_1kgp_countplots.pdf', dpi=300,
                    bbox_inches='tight')
```
